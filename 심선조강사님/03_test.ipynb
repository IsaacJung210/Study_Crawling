```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://finance.naver.com/marketindex/exchangeList.naver'
res = requests.get(url).text
soup = BeautifulSoup(res,'lxml')
result = soup.select('td.tit a')
name = []
name1 = []
price=[]
for item in result:
    v = item.string.strip().split()
    name.append(v[0].strip())
    name1.append(v[1].strip())

result = soup.select('td.sale')
for item in result:
    price.append(float(item.string.replace(',','')))
```

```python
import re
movie_list = []
for i in range(1,11):
    url = f'https://movie.naver.com/movie/point/af/list.naver?page={i}'
    res = urlopen(url).read()
    soup = BeautifulSoup(res,'html.parser')
    table = soup.select_one('table.list_netizen')
    for i,j in enumerate(table.select('tbody > tr')):
        # print(i)
        # print(j)
        for a,b in enumerate(j.select('td')):
            # print(a)
            # print(b)
            if a==1:
                # print(b)
                title = b.select_one('a').text
                # print(title)
                score = b.select_one('em').text
                # print(score)
                # print(b.text)
                cont = b.text.replace(title,'').replace('신고','')
                cont = re.sub('별점 - 총 10점 중[0-9]{1,2}','',cont)
                cont = cont.strip()
                # print(cont)
                movie_list.append((title,score,cont))
                dic = {'제목':title, '평점'}
movie_list

import json
with open('movielist.json','w') as f:
    json.dump(movie_list,f)
```
