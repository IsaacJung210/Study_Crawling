```python
import selenium, time, os, zipfile, shutil
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from urllib.request import urlretrieve
from tqdm import tqdm

#키워드 입력
keyword = input("수집할 이미지는 >>>")
if keyword == '':
    keyword ='벚꽃'
print('[접속중.....]')
driver = webdriver.Chrome()
driver.implicitly_wait(15)
url = f'https://search.naver.com/search.naver?where=image&sm=tab_jum&query={keyword}'
driver.get(url)

#화면 스크롤
print('[화면 스크롤.....]')
body = driver.find_element_by_css_selector('body')
for i in range(500):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)

links =[]
imgs = driver.find_elements_by_css_selector('img._image')
print(len(imgs))

#이미지 링크 추출
print('[이미지 링크 추출 중.....]')
for img in imgs:
    link = img.get_attribute('src')
    # print(link)
    if 'https' in link:
        links.append(link)
# print(links)

print('[정보 수집 완료 .....]')

#폴더 생성
print('[폴더 생성 중 .....]')
if not os.path.isdir(f'./{keyword}'):
    os.mkdir(f'./{keyword}')
    
# 이미지 다운로드
print('[이미지 다운로드 중 .....]')
for index,link in tqdm(enumerate(links),total=len(links)):
    start = link.rfind('.')
    end = link.rfind('&')
    file_type = link[start:end]
    # print(file_type)
    filename = f'./{keyword}/{keyword}{index:03d}{file_type}'
    # print(filename)
    urlretrieve(link,filename)
print('[이미지 다운로드 완료 .....]')

# #파일압축
# print('[이미지 압축 중 .....]')
# zip_file = zipfile.ZipFile(f'./{keyword}.zip','w')
# for image in os.listdir(f'./{keyword}'):
#     if image == '.ipynb_checkpoints':
#         pass
#     else:
#         # print(image)    
#         zip_file.write(f'./{keyword}/{image}',compress_type=zipfile.ZIP_DEFLATED)
# zip_file.close()
# print('[이미지 압축 완료 .....]')

# #폴더 삭제
# print('[이미지 폴더 삭제 중 .....]')
# shutil.rmtree(f'./{keyword}')
# print('[이미지 폴더 삭제완료 .....]')
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import os


def createDirectory(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print("Error: Failed to create the directory.")

def crawling_img(name):
    driver = webdriver.Chrome()
    driver.get("https://www.google.co.kr/imghp?hl=ko&tab=wi&authuser=0&ogbl")
    elem = driver.find_element_by_name("q")
    elem.send_keys(name)
    elem.send_keys(Keys.RETURN)

    #
    SCROLL_PAUSE_TIME = 1
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")  # 브라우저의 높이를 자바스크립트로 찾음
    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # 브라우저 끝까지 스크롤을 내림
        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)
        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            try:
                driver.find_element_by_css_selector(".mye4qd").click()
            except:
                break
        last_height = new_height

    imgs = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")
    dir = ".\idols" + "\\" + name

    createDirectory(dir) #폴더 생성
    count = 1
    for img in imgs:
        try:
            img.click()
            time.sleep(2)
            imgUrl = driver.find_element_by_xpath(
                '//*[@id="Sva75c"]/div/div/div[3]/div[2]/c-wiz/div/div[1]/div[1]/div[2]/div[1]/a/img').get_attribute(
                "src")
            path = "C:\\Users\\paqgl\\PycharmProjects\\pythonProject_crawling\\bs4\\idols\\" + name + "\\"
            urllib.request.urlretrieve(imgUrl, path + name + str(count) + ".jpg")
            count = count + 1
            if count >= 260:
                break
        except:
            pass
    driver.close()
src = input()
idols = [f"{src}"]

for idol in idols:
    crawling_img(idol)
```
